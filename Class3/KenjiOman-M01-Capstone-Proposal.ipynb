**Business Hypothetical:**
We have been funded by Maryland's Montgomery County local government to help encourage safe driving habits in their county.
```python
# To try to save the memory usage in this notebook, let's set the
# traffic violations variable to an empty dataframe, and then delete
# it (need to set it to an empty dataframe to get pandas' garbage
# collector to run)
tviolations_df = pd.DataFrame()
del tviolations_df
```

```python
# If we take a look at the column names, we see there are some that don't really
# make sense (non-intutive abbreviations)
weather_df.columns
```

```python
# So, let's rename the columns to make this a bit more intuitive
cols = {
    'DAPR': 'DaysInMultiPrecip',
    'MDPR': 'MultiPrecipTotal',
    'PRCP': 'Precipitation',
    'SNOW': 'Snow',
    'SNWD': 'SnowDepth',
    'TMAX': 'MaxTemp',
    'TMIN': 'MinTemp',
    'TOBS': 'ObservationTimeTemp',
    'WESD': 'GroundSnowWaterEquivalent',
    'WESF': 'SnowFallWaterEquivalent',
    'WT01': 'Fog',
    'WT03': 'Thunder',
    'WT04': 'SleetIceSnowPellets',
    'WT05': 'Hail',
    'WT06': 'GlazeRime',
    'WT11': 'HighWinds'
}

# All of these also have the _ATTRIBUTES suffix, so let's add conversions for them too
cols = {**cols, **{k + '_ATTRIBUTES': v + '_attr' for k, v in cols.items()}}

# Now, rename the columns
weather_df.rename(columns=cols, inplace=True)
weather_df.columns
```

```python
# Now, let's grab all the "attribute" columns, and see the unique values in them
# (note, I'm explicitly having value_counts show NaN values, but since there are
# a different number of unique values per column, some of the subsequent rows will
# show NaN for the attribute value, and its count. These can be ignored)
attributes = []
for col in weather_df.columns:
    if col[-5:] == '_attr':
        temp = weather_df[col].value_counts(dropna=False).reset_index()
        temp.rename(columns={col: 'values', 'index': col}, inplace=True)
        attributes.append(temp)
        
with pd.option_context('max_columns', 1000):
    display(pd.concat(attributes, axis=1))
```

```python
# Based on the values we have, and what they equate to via the documentation
# (at https://www1.ncdc.noaa.gov/pub/data/cdo/documentation/GHCND_documentation.pdf)
# it looks like there are a few values that might be useful, but the vast majority of them
# are not, so let's just drop these columns
weather_df.drop(columns=[col for col in weather_df.columns if col[-5:] == '_attr'], inplace=True)
```

```python
# Now, let's take a look at if there are other nominal columns (even
# if they are currently encoded as numbers)
attributes = []
for col in weather_df.columns:
    temp = weather_df[col].value_counts(dropna=False).reset_index()
    temp.rename(columns={col: 'values', 'index': col}, inplace=True)
    attributes.append(temp.head(8))
        
with pd.option_context('max_columns', 1000):
    display(pd.concat(attributes, axis=1))
```

```python
# From the above, we see some columns are most likely binary
# (they only have a value of NaN or 1), so let's set the NaN
# values to 0
bool_cols = ['Fog', 'Thunder', 'SleetIceSnowPellets', 'Hail', 'GlazeRime', 'HighWinds']
weather_df.loc[:, bool_cols] = weather_df.loc[:, bool_cols].fillna(0)
weather_df = weather_df.astype(dtype={x: int for x in bool_cols})
```

```python
# Also, drop the ObservationTimeTemp, as this isn't that useful
# (we would be able to extract the time, but they would have only
# been at either 8am or 10pm, for those measurements that have a
# time specified), and DaysInMultiPrecip only has one value,
# so isn't useful
weather_df.drop(columns=['ObservationTimeTemp', 'DaysInMultiPrecip'], inplace=True, errors='ignore')
```

```python
# Now, if we take a look at:
weather_df.groupby('STATION').mean()
```

```python
# We see the vast majority of stations don't actually provide any temperature data,
# so, let's just average over all the stations per date (the values
# across the county probably aren't going to be too different for
# those values from stations which have values)
temp = weather_df.drop(columns=['LATITUDE', 'LONGITUDE', 'ELEVATION'], errors='ignore').groupby(['DATE']).mean()
weather_df = temp.astype(dtype={x: bool for x in bool_cols}).astype(dtype={x: int for x in bool_cols})
weather_df.head()
```

```python
# Now, let's take the highly skewed data, and transform it to
# make it a little less skewed.  Also, since we have a lot of zeros,
# which don't fair well under a log transform, let's do a square
# root transform of these features
cols = ['Precipitation', 'Snow', 'SnowDepth', 'GroundSnowWaterEquivalent', 'SnowFallWaterEquivalent']
for col in cols:
    cleaned_weather_df.loc[:, col] = cleaned_weather_df.loc[:, col].apply(np.sqrt)
    
# And, now that we've transformed them, let's make some plots again
cleaned_weather_df[cols].plot(subplots=True, kind='box', figsize=(10,5), sym='c.', rot=45)
plt.show()
```
